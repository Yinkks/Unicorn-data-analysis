### INSIGHT 1
##### ROI

To determine the return on investment (ROI), i used both the initial investment and the final return. In this case, i approximated it by comparing the 'Funding' (total funding) to the 'Valuation' of each company. 

Once I've calculated the ROI for each company, I ranked them to find the ones with the highest ROI.

```python

df['ROI'] = ((df['Valuation'] - df['Funding']) / df['Funding'])*100
print(df['ROI'])
```

```python
## to sort the companies that had the biggest ROI
df_sorted = df.sort_values(by='ROI', ascending=False)
```

```python
top_10 = df_sorted.head(10)
print(top_10[['Company', 'ROI']])
```

```python
import matplotlib.pyplot as plt

# Assuming df is your dataframe
top_n = 10  # Number of companies you want to display, e.g., top 10
sorted_df = df.sort_values(by='ROI', ascending=False).head(top_n)

plt.figure(figsize=(12,7))

# Create bar chart
plt.bar(sorted_df['Company'], sorted_df['ROI'], color='RED')

# Title and labels
plt.title('Top Companies by Return on Investment (ROI)')
plt.xlabel('Company')
plt.ylabel('ROI (%)')
plt.xticks(rotation=45)  # Rotate company names for better visibility

# Display the plot
plt.tight_layout()
plt.show()
```

### i . observation:
Company Zapier  has  shown the most substantial returns on investments among all unicorn companies.

## How long does it usually take for a company to become a unicorn? Has it always been this way?

i calculated the time it took each company to become a unicorn by subtracting the 'Year Founded' from the year part of the 'Date Joined'.
Once i had  this duration for each company, i calculated the average time.
To see if it has always been this way, i grouped the companies by their 'Year Founded' or the year they became unicorns (from 'Date Joined') to see if the average time to become a unicorn has changed over the years.

```python
df['Year Joined'] = pd.to_datetime(df['Date Joined']).dt.year
```

```python
df['Time to Unicorn'] = df['Year Joined'] - df['Year Founded']
```

```python
average_time = df['Time to Unicorn'].mean()
print(f"Average time to become a unicorn: {average_time} years")
```

```python
grouped = df.groupby('Year Founded')['Time to Unicorn'].mean()

print(grouped)
```

```python
import matplotlib.pyplot as plt
```

```python
grouped.plot()
plt.ylabel('Average Time to Unicorn (years)')
plt.xlabel('Year Founded')
plt.title('Trend of Time to Become a Unicorn Over the Years')
plt.show()
```

## observation and recommendation for insight 2

### i . observation:
On average, it takes 7  years for companies to achieve unicorn status.

## Which investors have funded the most unicorns?

1. I'll  split the 'Select Investors' column into a list of individual investors for each company.
 2. count how often each investor appears in the 'Investors' lists. One way to do this is to flatten the lists and then use a Counter to count occurrences.
 3. Sort the Investors by Count
 4. visualize
 



```python
df['Investors'] = df['Select Investors'].str.split(',')
```

```python
from collections import Counter

all_investors = [investor.strip() for sublist in df['Investors'].dropna() for investor in sublist]
investor_counts = Counter(all_investors)
```

```python
sorted_investors = sorted(investor_counts.items(), key=lambda x: x[1], reverse=True)
```

```python
# Display top 10 investors
for investor, count in sorted_investors[:10]:
    print(f"{investor}: {count} unicorns")
```

```python
import matplotlib.pyplot as plt

# Extract top 10 investors and their counts
top_investors = [x[0] for x in sorted_investors[:10]]
top_counts = [x[1] for x in sorted_investors[:10]]

# Plotting
plt.figure(figsize=(12, 7))
plt.barh(top_investors, top_counts, color='purple')
plt.xlabel('Number of Unicorns Funded')
plt.title('Top 10 Investors Funding the Most Unicorns')
plt.gca().invert_yaxis()  # To display the top investor at the top
plt.show()
```

## observation and recommendation for insight 4 

### i . observation:
Investors Accel, Tiger Global Management, and Adreessen Horowitz  are the top players in funding unicorn companies, having a diverse portfolio across multiple sectors.

These investors often engage early in a startup's lifecycle, contributing not just funds but also mentorship and strategic guidance.

### ii. recommendation: 
Strategic Partnerships: Startups should aim to secure investments from these top players as it might provide them with better networking opportunities, mentorship, and credibility in the market.

Diverse Portfolio: Investors can look into diversifying their portfolios across various sectors, as seen with the top investors.

## INSIGHT 7 
## How 'Valuation' has trended over time (are unicorns being valued more highly in recent years?).

To understand how 'Valuation' has trended over time, I will analyze the valuation based on the 'Date Joined' column (the date when companies reached $1 billion valuation). Here's a step-by-step process:

Data Preparation:

Ensured there are no missing values in the 'Valuation' and 'Date Joined' columns.
Converted  the 'Date Joined' column to a datetime .

Grouping Data by Year:

To see the trend over time, i grouped the data by year.

```python
import pandas as pd
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
```

```python
print(df['Date Joined'].dtypes)
```

```python
# Extracting year from 'Date Joined' and creating a new column 'Year Joined'
df['Year Joined'] = df['Date Joined'].dt.year

# Calculating the average valuation for each year
yearly_valuation = df.groupby('Year Joined')['Valuation'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(yearly_valuation['Year Joined'], yearly_valuation['Valuation'], marker='o', linestyle='-')
plt.title('Trend of Average Valuation Over Time')
plt.xlabel('Year')
plt.ylabel('Average Valuation (in billions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## observation and recommendation for insight 7 

### i . observation:the line went up in between 2008 , reached an high peak by  2012 then started trending downward , it suggests that the that unicorns are being valued more lower  in recent years.

### ii. recommendation: "The decreasing trend might be indicative of market saturation, changing investor sentiment, or a more cautious approach to valuing unicorns. Stakeholders need to understand the underlying reasons and strategize accordingly, taking into consideration the processes that they adopted during the peak years."
